```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("income.csv")
df.head()
df
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
km = KMeans(3)
```

```python
km
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head() #first five
df
```

```python
plt.scatter(df["Age"], df["Income($)"],c = df["cluster"],cmap="rainbow")
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

Scaling the age from 0-1. This process makes the axis narrower, since x-axis is narrower than the y-axis.

```python
#the scale of x-axis is narrower than that of y-axis
#transforming the income column
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python
#transforming the age column
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
#transformed plot but still the error is there
plt.scatter(df["Age"], df["Income($)"],c = df["cluster"],cmap="rainbow")
```

```python
km = KMeans(3)
km
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
plt.scatter(df["Age"], df["Income($)"],c = df["cluster"],cmap="rainbow")
#x-axis--->km.cluster_centers_[:,0]
#y-axis--->km.cluster_centers_[:,1]
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
```

```python
sse = []#create an empty array
#//k_rng = range(1,10)
k_rng = range(1,10) # //iterate from 1 to 10

for k in k_rng:
  #in each iteration, create a new model with clusters = k
  km = KMeans(n_clusters = k)
  #create a new data frame excluding the name col rather than passing the column names every time 
    #km.fir(<dataframe>)
  km.fit(df[['Age','Income($)']])
  #inertia will give the sse
  sse.append(km.inertia_)
```

```python
#sum of squared errors
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
